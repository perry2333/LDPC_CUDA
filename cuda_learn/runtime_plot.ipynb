```python
import re
import matplotlib.pyplot as plt

# 读取文件内容
with open('test_wrap_result.txt', 'r') as f:
    lines = f.readlines()

# 提取线程数和时间
threads = []
times = []

for line in lines:
    match = re.search(r'Avg time for 1 blocks x (\d+) threads: ([\d.]+) ms', line)
    if match:
        threads.append(int(match.group(1)))
        times.append(float(match.group(2)))

# 绘图
plt.figure(figsize=(10, 6))
plt.plot(threads, times, marker='.')
plt.xlabel('Threads per Block')
plt.ylabel('Average Execution Time (ms)')
plt.title('CUDA Kernel Execution Time vs Threads per Block')
plt.grid(True)
plt.tight_layout()
plt.show()
```
